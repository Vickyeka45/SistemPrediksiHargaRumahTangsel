```python
import pandas as pd

# Membaca data dari file excel
df = pd.read_excel('Data KTTA Terbaru Fix.xlsx')

# Menghitung modus dari kolom sertifikat
modus_sertifikat = df['Sertifikat_'].mode()[0]

# Mengisi nilai-nilai yang hilang di kolom sertifikat dengan nilai modus
df_sertifikat_filled = df['Sertifikat_'].fillna(modus_sertifikat)
df.update(df_sertifikat_filled)

# Menghitung modus dari kolom Daya Listrik (Watt)
modus_DayaListrik = df['Daya Listrik (Watt)'].mode()[0]

# Mengisi nilai-nilai yang hilang di kolom Daya Listrik (Watt) dengan nilai modus
df_DayaListrik_filled = df['Daya Listrik (Watt)'].fillna(modus_DayaListrik)
df.update(df_DayaListrik_filled)

# Menghitung modus dari kolom Jumlah Lantai
modus_JumlahLantai = df['Jumlah Lantai'].mode()[0]

# Mengisi nilai-nilai yang hilang di kolom Jumlah Lantai dengan nilai modus
df_JumlahLantai_filled = df['Jumlah Lantai'].fillna(modus_JumlahLantai)
df.update(df_JumlahLantai_filled)

# Menghitung modus dari kolom Carport
modus_Carport = df['Carport'].mode()[0]

# Mengisi nilai-nilai yang hilang di kolom Carport dengan nilai modus
df_Carport_filled = df['Carport'].fillna(modus_Carport)
df.update(df_Carport_filled)

# Menghitung modus dari kolom Garasi
modus_Garasi = df['Garasi'].mode()[0]

# Mengisi nilai-nilai yang hilang di kolom Garasi dengan nilai modus
df_Garasi_filled = df['Garasi'].fillna(modus_Garasi)
df.update(df_Garasi_filled)

# Menghitung modus dari kolom Interior
modus_Interior = df['Interior'].mode()[0]

# Mengisi nilai-nilai yang hilang di kolom Garasi dengan nilai modus
df_Interior_filled = df['Interior'].fillna(modus_Interior)
df.update(df_Interior_filled)

# Menyimpan perubahan kembali ke file excel jika diperlukan
df.to_csv('Data KTTA.csv', index=False)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('Data Olahan KTTA_.csv')

# Variabel Independen
X = data.drop(['web-scraper-order','web-scraper-start-url','link','link-href','Lokasi','Nama Perumahan','Orientasi Bangunan','Harga'], axis=1)
X.head()

# Variabel Dependen
y=data["Harga"]
y.head()

# Data Testing dan Training
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.5,
                                                    random_state=1234)

# Proporsi Data Testing dan Training
print(X_train.shape)
print(X_test.shape)

# Create Random Forest Regressor with n_estimators=1000 and max_features="sqrt"
regressor = RandomForestRegressor(n_estimators=1000,
                                   max_features="sqrt",
                                  random_state=1234)

# Train the model
regressor.fit(X_train, y_train)

# Make predictions on the test set
y_pred=regressor.predict(X_test)


#Nilai Koefiisien Determinasi Data Testing
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
y_pred2=regressor.predict(X_train)
r2_score(y_train, y_pred2)
# Coefficient of Determination (R^2) for Testing Data
print('R^2 Score (Testing Data):', r2_score(y_test, y_pred))

# Coefficient of Determination (R^2) for Training Data
y_pred_train = regressor.predict(X_train)
print('R^2 Score (Training Data):', r2_score(y_train, y_pred_train))

#Perhitungan Akurasi
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test,
y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test,
y_pred))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('MAPE :', np.mean(np.abs((y_test - y_pred) / y_test))*100)
print ('Nilai akurasi :', (1-(np.mean(np.abs((y_test - y_pred) /
y_test))))*100)


# Combine actual and predicted prices for both training and testing data
actual_prices = np.concatenate([y_train, y_test])
predicted_prices = np.concatenate([y_pred_train, y_pred])

# Create a new DataFrame that contains the actual and predicted prices
df = pd.DataFrame({'Actual Price': actual_prices, 'Predicted Price': predicted_prices})

# Display the first few rows of the DataFrame
print(df.head())

# Create a line plot of actual vs predicted prices
plt.plot(df.index, df['Actual Price'], linestyle='--', label='Actual Price', color='blue')
plt.plot(df.index, df['Predicted Price'], label='Predicted Price', color='red')

# Add axis labels and a title
plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Actual Price vs Estimated Price Using The Random Forest Method')
plt.legend()

# Display the plot
plt.show()


# Combine actual and predicted prices for both training and testing data
actual_prices = np.concatenate([y_train, y_test])
predicted_prices = np.concatenate([y_pred_train, y_pred])

# Create a new DataFrame that contains the actual and predicted prices
df = pd.DataFrame({'Actual Price': actual_prices, 'Predicted Price': predicted_prices})

# Display the first few rows of the DataFrame
print(df.head())

# Create a scatter plot of actual vs predicted prices
plt.scatter(df['Actual Price'], df['Predicted Price'])

# Add a line of best fit to the plot
m, b = np.polyfit(df['Actual Price'], df['Predicted Price'], 1)
plt.plot(df['Actual Price'], m*df['Actual Price'] + b, '-r')

# Add axis labels and a title
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Estimated Price Using The Random Forest Method')

# Display the plot
plt.show()

# Feature Importance
# Feature importances into a dataframe
features = list(X_train.columns)
feature_importances = pd.DataFrame({'feature': features,
'importance': regressor.feature_importances_})
feature_importances .plot(x ='feature', y='importance', kind =
'barh', color="lime")
print(feature_importances)
```
